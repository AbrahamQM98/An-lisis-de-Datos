```python
import pandas as pd
import matplotlib.pyplot as plt

dfr=pd.read_csv('/casasboston.csv')
data=pd.DataFrame(dfr)
data
```

```python
df=data[["RM","CRIM", "MEDV","TOWN","CHAS"]]
```

```python
df=data.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"
})
df
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(grid=True, legend=True,bins=12, alpha=0.9)
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN", alpha=0.5)
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
